```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b0 = 9
b1 = 6
```

```python
x =  np.random.random(100)
y = b0 + b1 * x
print(y.shape)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
B1 = np.sum ((y - np.mean(y)) * x) /np.sum ((x - np.mean(x))*x)
print(B1)
```

```python
B0 = np.mean(y) - B1 * np.mean(x)
print(B0)
```

```python
yn = y + np.random.normal(0,2,100)
print(yn)
plt.scatter(x,y)
plt.scatter(x,yn)
plt.show()
```

```python
noisyB1 = np.sum ((yn - np.mean(yn)) * x) /np.sum ((x - np.mean(x))*x)
print(noisyB1)
```

```python
noisyB0 = np.mean(yn) - B1 * np.mean(x)
print(noisyB0)
```

```python
yest = noisyB0 + noisyB1 * x
plt.scatter(x,y)
plt.scatter(x,yn)
plt.scatter(x,yest)
plt.show()
```

## HW
### This code performs the following:

### Residuals: Calculates the difference between actual (y) and predicted (yest) values.
### MAE: Averages the absolute residuals.
### MSE: Averages the squared residuals.

```python
# Calculate residuals (errors)
error = y - yest

# Calculate Mean Absolute Error (MAE)
mae = np.mean(np.abs(error))

# Calculate Mean Squared Error (MSE)
mse = np.mean(error**2)

# Print the results
print("Residuals (First 5):", error[:5])
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
```
