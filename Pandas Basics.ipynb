```python
numC = 0
for index,row in x_test.iterrows():
    d = x_train.apply(lambda r: distances(row,r), axis = 1)
    mini = np.argmin(d)
    print(mini)
    if(y_train.iloc[mini] == y_test.loc[index]):
        numC = numC+1
print(numC)
```
