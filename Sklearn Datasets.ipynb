## Difference:
### load_* Datasets: Small datasets bundled with scikit-learn.
### fetch_* Datasets: Larger datasets that are fetched from external resources (often from the internet).

#### Load Datasets

```python
# 1.Import Libraries
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# 2.Load Two Datasets
# Iris Dataset – A multi-class classification dataset.
# Wine Dataset – Another multi-class classification dataset.
```

```python
iris = datasets.load_iris()
wine = datasets.load_wine()
```

```python
# Summary of the Datasets
```

```python
# Iris Dataset
print("Iris Dataset Description:\n", iris.DESCR[:500], "\n...")
print("Number of Samples:", iris.data.shape[0])
print("Number of Features:", iris.data.shape[1])
print("Number of Classes:", len(np.unique(iris.target)))
print("\nClass Labels:", iris.target_names)

# Wine Dataset
print("\nWine Dataset Description:\n", wine.DESCR[:500], "\n...")
print("Number of Samples:", wine.data.shape[0])
print("Number of Features:", wine.data.shape[1])
print("Number of Classes:", len(np.unique(wine.target)))
print("\nClass Labels:", wine.target_names)
```

#### Fetch Datasets

```python
# 1.Fetch Two Datasets
```

```python
from sklearn.datasets import fetch_openml, fetch_california_housing

# Fetch datasets
mnist = fetch_openml('mnist_784', version=1)  
california_housing = fetch_california_housing() 
```

```python
# 2.Explore Fetched Datasets
```

```python
# Display MNIST dataset details
print("MNIST Dataset")
print("Number of samples:", mnist.data.shape[0])
print("Number of features:", mnist.data.shape[1])
print("Number of classes:", len(np.unique(mnist.target)))
print("\nClasses:", np.unique(mnist.target))
```

```python
# Display California housing dataset details
print("\nCalifornia Housing Dataset")
print("Number of samples:", california_housing.data.shape[0])
print("Number of features:", california_housing.data.shape[1])
print("Target description:", california_housing.DESCR.split('\n')[0])
```

```python
# 3.Convert to Pandas DataFrame
```

```python
# Convert fetched datasets to DataFrames
mnist_df = pd.DataFrame(mnist.data)
mnist_df['target'] = mnist.target

california_df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
california_df['target'] = california_housing.target
```

```python
# 4.Check the Structure and Summary of DataFrames
```

```python
# Check the first few rows of each DataFrame
print("MNIST Dataset (First 5 Rows):")
print(mnist_df.head())

print("\nCalifornia Housing Dataset (First 5 Rows):")
print(california_df.head())

# Check for missing values
print("\nMissing Values in MNIST Dataset:")
print(mnist_df.isnull().sum())

print("\nMissing Values in California Housing Dataset:")
print(california_df.isnull().sum())

# Get basic statistics
print("\nMNIST Dataset Summary:")
print(mnist_df.describe())

print("\nCalifornia Housing Dataset Summary:")
print(california_df.describe())
```

```python
# 5.Visualize the Datasets
```

```python
# MNIST Dataset (Digit Distribution)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot class distribution in MNIST
sns.countplot(x='target', data=mnist_df)
plt.title("Distribution of Digits in MNIST Dataset")
plt.show()
```

```python
# California Housing (Histogram of Target Values)
```

```python
# Plot histogram for the target (house value)
plt.hist(california_df['target'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Median House Values")
plt.xlabel("Median House Value ($100,000)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 6.Correlation Analysis (California Housing)
```

```python
# Calculate correlation matrix
correlation_matrix = california_df.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (California Housing)")
plt.show()
```

```python
# Plot one of the images from the MNIST dataset
import matplotlib.pyplot as plt

sample_image = mnist.data.iloc[0].values.reshape(28, 28)
plt.imshow(sample_image, cmap='gray')
plt.title(f"Label: {mnist.target.iloc[0]}")
plt.show()
```
