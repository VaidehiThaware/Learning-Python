```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img3 = cv2.imread("C:\\Users\\Vaidehi Thaware\\Downloads\\MainAfter.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt


a = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
print(a)

newimages = np.empty_like(img3)
f = 9

for r in range(1, img3.shape[0] - 1):
    for c in range(1, img3.shape[1] - 1):
       
        B = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 0] * a) // f
        newimages[r, c, 0] = B
        newimages[r, c, 1] = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 1] * a) // f
        newimages[r, c, 2] = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 2] * a) // f
    if r % 200 == 0:
        print("done with rowR", r)

plt.imshow(newimages)


a = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
print(a)

newimage2 = np.empty_like(img3)
f = 9

for r in range(1, img3.shape[0] - 1):
    for c in range(1, img3.shape[1] - 1):
        B = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 0] * a) // f
        newimage2[r, c, 0] = B
        newimage2[r, c, 1] = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 1] * a) // f
        newimage2[r, c, 2] = np.sum(img3[r - 1:r + 2, c - 1:c + 2, 2] * a) // f
    if r % 200 == 0:
        print("done with rowR", r)

plt.imshow(newimage2)


fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(newimages)
ax[0].set_title("Filtered Image 1")
ax[1].imshow(newimage2)
ax[1].set_title("Filtered Image 2")
plt.show()
```
