```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
sns.relplot(x = "passengers",y = "month",data = a)
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="month", y="passengers", hue="year", data=a)
```

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y="tip",data=b,kind = "line")
```

```python
sns.catplot(x = "day", y = "total_bill",data = b)
```

```python
sns.catplot(x = "day", y = "total_bill",kind = "violin",data = b)
```

```python
sns.catplot(x = "day", y = "total_bill",kind = "boxen",data = b)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5, 6, 7],
    'Grocery': [30, 80, 45, 23, 51, 46, 76],
    'Clothes': [13, 40, 34, 23, 54, 67, 98],
    'Utensils': [12, 32, 27, 40, 45, 50, 37]
})

y = pd.DataFrame({
    'Day': [8, 9, 10, 11, 12, 13, 14],
    'Grocery': [30, 80, 45, 23, 51, 46, 76],
    'Clothes': [13, 40, 34, 23, 54, 67, 98],
    'Utensils': [12, 32, 27, 40, 45, 50, 37]
})

mean = [0, 1]
cov = [[1, 0.5], [0.5, 1]]
data = np.random.multivariate_normal(mean, cov, 200)
x_data = data[:, 0]
y_data = data[:, 1]

with sns.axes_style("white"):
    sns.jointplot(x=x_data, y=y_data, kind="kde", color="b")

plt.show()
```
