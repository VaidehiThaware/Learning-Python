```python
img2 = np.zeros((img.shape[0]//2,img.shape[1]//2,3), dtype = 'uint8')
```

```python
rows = img.shape[0]
cols = img.shape[1]
for r in range(rows):
    if (r%2 == 0):
        for c in range(cols):
            if c%2 == 0:
                img2[r//2,c//2] = img[r,c]
```
